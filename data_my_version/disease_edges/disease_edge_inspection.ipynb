```python
hyperedge_df = pd.DataFrame(hyperedges).T
```

```python
hyperedge_df.info()
```

```python
hyperedge_df[:100]
```

```python
tuple_rows = []
arities = []

for _, row in hyperedge_df.iterrows():
    hyperedge = (
        ["DISEASE:" + row["disease"]] +
        ["PARENT_DISEASE:" + str(x) for x in row["parents"]] +
        ["ASSOCIATED_DISEASE:" + str(x) for x in row["associated_diseases"]] +
        ["COMORBID_DISEASE:" + str(x) for x in row["comorbidity_diseases"]] +
        ["DRUG:" + str(x) for x in row["drugs"]] +
        ["GENE:" + str(x) for x in row["genes"]] +
        ["ORGANISM:" + str(x) for x in row["organisms"]] +
        ["PHENOTYPE:" + str(x) for x in row["phenotypes"]]
    )

    tuple_rows.append({
        "relation": "disease_context",
        "nodes": hyperedge,
        "arities":len(hyperedge)
    })
```

```python
tuple_df = pd.DataFrame(tuple_rows)
```
